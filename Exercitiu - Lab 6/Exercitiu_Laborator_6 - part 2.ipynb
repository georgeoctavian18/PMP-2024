```python
import scipy.stats as stats
import arviz as az

y = 180
n = 10

alpha_prior = 18 
beta_prior = 1

alpha_posterior = alpha_prior + y
beta_posterior = beta_prior + n

posterior_dist = stats.gamma(a=alpha_posterior, scale=1/beta_posterior)

mode_lambda = (alpha_posterior - 1) / beta_posterior if alpha_posterior > 1 else 0

posterior_samples = posterior_dist.rvs(10000)
hdi_interval = az.hdi(posterior_samples, hdi_prob=0.94)

print("Parametrii distributiei a posteriori Gamma:")
print(f"alpha_posterior = {alpha_posterior}")
print(f"beta_posterior = {beta_posterior}\n")
print(f"Modulul (valoarea cea mai probabila) pentru 位 = {mode_lambda}")
print(f"Intervalul HDI de 94% pentru 位: {hdi_interval}")

az.plot_posterior(posterior_samples, hdi_prob=0.94)
```
